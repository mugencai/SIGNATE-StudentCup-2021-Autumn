## 1.静態図を出力する

```python
import matplotlib.pyplot as plt
import numpy as np


fig, ax = plt.subplots(figsize=(8,4),dpi=200) 
plt.axis('off')


#legend
ax.plot([2], [12],"ob", label="訓練データ" ) 
ax.plot([4], [12],"og", label="検証データ" ) 
ax.plot([6], [12],"or", label="テストデータ" ) 
ax.legend(frameon=False,loc =8,ncol= 3,prop={"family":"Yu Gothic"} )

#text
ax.text(25.5, 11.8, 'time')


for x in range(12,24):
    #右矢印を書く
    ax.plot(np.array(range(0,24+2)), np.array([24-x]*(24+2)), "o-k",marker=5, markevery=[24+1])
    #中空円を書く
    ax.plot(np.array(range(1,24+2)), np.array([24-x]*(24+1)), "ok", markevery=range(24),markerfacecolor='white')
    
    #前x個の円をブルー色に染める
    ax.plot(np.array(range(1,x+1)), np.array([24-x]*(x)), "ok", mfc='b',mec="b", markevery=range(x))
    
    #第x+1個の円を赤色に染める
    ax.plot([x+1], [24-x],"or") 
    
    #図（ブルーと赤）を保存
    ax.set_xticks([-2,0,5,10,15,20,25,27]) 
    ax.set_yticks([-1,0,2,4,6,8,10,12,14]) 
    fig.savefig('./pic/%s_1.jpg'%(x))
    
    #第x個の円を緑色に染める
    ax.plot([x], [24-x],"og")
    
    #図（ブルー、赤と緑）を保存
    ax.set_xticks([-2,0,5,10,15,20,25,27]) 
    ax.set_yticks([-1,0,2,4,6,8,10,12,14]) 
    fig.savefig('./pic/%s_2.jpg'%(x))
```

## 2.imageio.mimsaveと先ほど出力された図でgifを作る

```python
import os
import imageio
 
def create_gif(image_list, gif_name):
    frames = []
    for image_name in image_list:
        if image_name.endswith(".jpg"):
            print(image_name)
            frames.append(imageio.imread(image_name))
    imageio.mimsave(gif_name, frames, 'GIF', duration = [3.5]+[3,1.5]*2+[3.5,5]+[0.6]*3)
    return
    
def main():
    path=r'./pic/'
    image_list=[path+img for img in[ "20_2.jpg", "21_1.jpg","21_2.jpg", "22_1.jpg", "22_2.jpg", "23_1.jpg","23_2.jpg","22_2.jpg","21_2.jpg","20_2.jpg",]]
    gif_name = 'slide_wind.gif'
    create_gif(image_list, gif_name)
    
main()
```
